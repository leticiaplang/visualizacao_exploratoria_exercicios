## Dataset

Variables:
- Name: pokemon's name
- Type 1: pokemon type 1 (evolution)
- Type 2: pokemon type 2 (evolution)
- Total: total pokemon numbers?
- HP: life score
- Attack: attack score
- Defense: defense score
- Sp. Atk: special attack score
- Sp. Def: special defense score
- Speed: agile score
- Stage: evolutionary state
- Legendary: is this a legendary pokemon?


## Preparing notebook

```python
# imports
%matplotlib inline
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
# Checking the DF
df = pd.read_csv('pokemon.csv', encoding='latin-1', index_col=0)
df.head()
```

## Color palette

[Bulbapedia](https://bulbapedia.bulbagarden.net/wiki/Category:Type_color_templates)

```python
pokemon_colors = [
    '#78C850',  # Grass
    '#F08030',  # Fire
    '#6890F0',  # Water
    '#A8B820',  # Bug
    '#A8A878',  # Normal
    '#A040A0',  # Poison
    '#F8D030',  # Electric
    '#E0C068',  # Ground
    '#EE99AC',  # Fairy
    '#C03028',  # Fighting
    '#F85888',  # Psychic
    '#B8A038',  # Rock
    '#705898',  # Ghost
    '#98D8D8',  # Ice
    '#7038F8'  # Dragon
]
```

```python
plt.figure(figsize=(12, 6))
sns.violinplot(x='Type 1', y='Attack', data=df, palette=pokemon_colors)
```

## Task 05
* Create a swarmplot using the color palette

```python
plt.figure(figsize=(12, 6))
sns.swarmplot(x='Type 1', y='Attack', data=df, palette=pokemon_colors)
```

```python
# Expected result
```

## Graphics overlay


```python
plt.figure(figsize=(10,6))
 
# Creating violinplot
sns.violinplot(x='Type 1', y='Attack', data=df, inner=None, palette=pokemon_colors)

# Criating swarmplot
sns.swarmplot(x='Type 1', y='Attack', data=df, color='k', alpha=0.7)
 
# Defining title
plt.title('Ataque por tipo de Pokemon')
```
